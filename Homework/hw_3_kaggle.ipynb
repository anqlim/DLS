Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
for col in num_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')
    data[col] = data[col].fillna(data[col].median())

# Separate features and target
X = data.drop(target_col, axis=1)
y = data[target_col]

# Convert 'Churn' target variable to numerical (0 and 1) if not already
# Assuming 'Yes' for 1 and 'No' for 0 based on typical churn datasets
if y.dtype == 'object':
    y = y.map({'Yes': 1, 'No': 0})

# Split data into training and validation sets
X_train, X_test_val, y_train, y_test_val = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Define preprocessing steps
# Numerical features: StandardScaler
# Categorical features: OneHotEncoder (handle_unknown='ignore' to prevent errors on unseen categories)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])

# Create a pipeline with preprocessing and Logistic Regression
pipeline = make_pipeline(preprocessor, LogisticRegression(solver='liblinear', random_state=42))

# Define hyperparameter grid for C
param_grid = {
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]
}

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5, # 5-fold cross-validation
    scoring='roc_auc', # Use ROC-AUC as the scoring metric
    verbose=1,
    n_jobs=-1 # Use all available CPU cores
)

# Train the GridSearchCV model
grid_search.fit(X_train, y_train)

# Print the best score and parameters
best_score = grid_search.best_score_
best_params = grid_search.best_params_

print(f"Best ROC-AUC score from GridSearchCV: {best_score:.4f}")
print(f"Best parameters for Logistic Regression: {best_params}")

# Evaluate the best model on the validation set
best_model_lr = grid_search.best_estimator_
y_pred_proba_val = best_model_lr.predict_proba(X_test_val)[:, 1]
roc_auc_val = roc_auc_score(y_test_val, y_pred_proba_val)
print(f"ROC-AUC on validation set with best Logistic Regression: {roc_auc_val:.4f}")
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

for col in cat_cols:
    X_train[col] = X_train[col].astype('object')
    X_test_val[col] = X_test_val[col].astype('object')

cat_default = CatBoostClassifier(random_state=42, cat_features=cat_cols, verbose=0)
cat_default.fit(X_train, y_train)

y_pred_proba_default = cat_default.predict_proba(X_test_val)[:, 1]
roc_auc_default = roc_auc_score(y_test_val, y_pred_proba_default)
print(f"ROC-AUC on validation set with default CatBoost: {roc_auc_default:.4f}")

# Hyperparameter tuning for CatBoost
param_grid_cat = {
    'iterations': [100, 200, 300, 500],  # ← было n_estimators
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'depth': [4, 6, 8]
}

cat_model = CatBoostClassifier(random_state=42, verbose=0, early_stopping_rounds=50)  # ← убрали cat_features=cat_cols

grid_search_cat = GridSearchCV(
    estimator=cat_model,
    param_grid=param_grid_cat,
    cv=3,
    scoring='roc_auc',
    verbose=1,
    n_jobs=-1
)

# Train the GridSearchCV model
grid_search_cat.fit(X_train, y_train)

# Print the best score and parameters found by GridSearchCV
best_score_cat = grid_search_cat.best_score_
best_params_cat = grid_search_cat.best_params_

print(f"\nBest ROC-AUC score for CatBoost: {best_score_cat:.4f}")
print(f"Best parameters for CatBoost: {best_params_cat}")

# Evaluate the best CatBoost model on the validation set
best_model_cat = grid_search_cat.best_estimator_
y_pred_proba_cat_val = best_model_cat.predict_proba(X_test_val)[:, 1]
roc_auc_cat_val = roc_auc_score(y_test_val, y_pred_proba_cat_val)
print(f"ROC-AUC on validation set with best CatBoost: {roc_auc_cat_val:.4f}")
```
